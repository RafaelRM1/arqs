Usando arquivos no Colab:

1. Faça o upload do arquivo no Colab usando a aba da esquerda.

2. Abra o arquivo de entrada no Colab usando o comando open com a opção read (r)

3. Leia o arquivo linha a linha com readline.

4. Para gravação abra o arquivo de saída com opção write (w).

5. Grave no arquivo de saída usando o comando write.

6. Feche o arquivo com o comando close.




```python
with open('teste1.txt', 'r') as f:
  n = int(f.readline())
  print(n)
  l = []
  for i in range(n):
    line = f.readline()
    l.append(line)
    print(line)
```

```python
with open('saida.txt', 'w') as f:
  for x in l:
    f.write(x)
  f.close()
```

```python
#eita
texto = input()
palavra = input()

# Contando ocorrências da palavra
ocorrencias = texto.count(palavra)

with open("remis.txt","a") as arq:
  arq.write(f"A palavra '{palavra}' aparece {ocorrencias} vezes no texto.")
```

```python
arq = open("remis.txt","r")
x = arq.read()
print(x)
```

```python
arq = open("meu_arquivo2.txt","w+")

linhasParaOArquivo = ["linha 1","linha 2","linha 3", \
"linha 4","linha 5"]

for lnh in linhasParaOArquivo:
    arq.write(lnh)
    arq.write("\n")
arq.close()
```
